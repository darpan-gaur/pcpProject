```python
import numpy as np
import matplotlib.pyplot as plt

averageTime = []
averageError = []
gridSize = [100, 150, 200, 250, 300, 350, 400]
```

```python
l1 = ['dynamicRow.txt', 'dynamicRowBatch.txt', 'dynamicRowChunk.txt', 'dynamicRowChunkBatch.txt']

for name in l1:
    with open(name, 'r') as f:
        data = f.read().split('\n')
    
    data = data[1:]

    avgTime = []
    avgError = []
    i = 0
    while (i<len(data)-1):
        i += 1
        avgTime.append(np.mean([float(data[j]) for j in range(i, i+15, 3)]))
        i += 1
        avgError.append(np.mean([float(data[j]) for j in range(i, i+15, 3)]))
        i += 13
    averageTime.append(avgTime)
    averageError.append(avgError)

plt.plot(gridSize, averageTime[0], 'o-', label='dynamicRow')
plt.plot(gridSize, averageTime[1], 'o-', label='dynamicRowBatch')
plt.plot(gridSize, averageTime[2], 'o-', label='dynamicRowChunk')
plt.plot(gridSize, averageTime[3], 'o-', label='dynamicRowChunkBatch')
plt.xlabel('Grid Size')
plt.ylabel('Average Time')
plt.title('Average Time vs Grid Size')
plt.legend()
plt.show()

plt.plot(gridSize, averageError[0], 'o-', label='dynamicRow')
plt.plot(gridSize, averageError[1], 'o-', label='dynamicRowBatch')
plt.plot(gridSize, averageError[2], 'o-', label='dynamicRowChunk')
plt.plot(gridSize, averageError[3], 'o-', label='dynamicRowChunkBatch')
plt.xlabel('Grid Size')
plt.ylabel('Average Error')
plt.title('Average Error vs Grid Size')
plt.legend()
```

```python
l2 = ['groupRow.txt', 'groupRowBarrierFree.txt', 'groupRowBatch.txt']
l22 = ['oneRow.txt', 'oneRowBarrierFree.txt', 'oneRowBatch.txt']

for name in l2:
    with open(name, 'r') as f:
        data = f.read().split('\n')
    
    data = data[1:]

    avgTime = []
    avgError = []
    i = 0
    while (i<len(data)-1):
        i += 1
        avgTime.append(np.mean([float(data[j]) for j in range(i, i+1, 3)]))
        i += 1
        avgError.append(np.mean([float(data[j]) for j in range(i, i+1, 3)]))
        i += 1
    averageTime.append(avgTime)
    averageError.append(avgError)

for name in l22:
    with open(name, 'r') as f:
        data = f.read().split('\n')
    
    data = data[1:]

    avgTime = []
    avgError = []
    i = 0
    while (i<len(data)-1):
        i += 1
        avgTime.append(np.mean([float(data[j]) for j in range(i, i+13, 3)]))
        i += 1
        avgError.append(np.mean([float(data[j]) for j in range(i, i+13, 3)]))
        i += 13
    averageTime.append(avgTime)
    averageError.append(avgError)

plt.plot(gridSize, averageTime[4], 'o-', label='groupRow')
plt.plot(gridSize, averageTime[5], 'o-', label='groupRowBarrierFree')
plt.plot(gridSize, averageTime[6], 'o-', label='groupRowBatch')
plt.plot(gridSize, averageTime[7], 'o-', label='oneRow')
plt.plot(gridSize, averageTime[8], 'o-', label='oneRowBarrierFree')
plt.plot(gridSize, averageTime[9], 'o-', label='oneRowBatch')
plt.xlabel('Grid Size')
plt.ylabel('Average Time')
plt.title('Average Time vs Grid Size')
plt.legend()
plt.show()

plt.plot(gridSize, averageError[4], 'o-', label='groupRow')
plt.plot(gridSize, averageError[5], 'o-', label='groupRowBarrierFree')
plt.plot(gridSize, averageError[6], 'o-', label='groupRowBatch')
plt.plot(gridSize, averageError[7], 'o-', label='oneRow')
plt.plot(gridSize, averageError[8], 'o-', label='oneRowBarrierFree')
plt.plot(gridSize, averageError[9], 'o-', label='oneRowBatch')
plt.xlabel('Grid Size')
plt.ylabel('Average Error')
plt.title('Average Error vs Grid Size')
plt.legend()
plt.show()
```
