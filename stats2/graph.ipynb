```python
import numpy as np
import matplotlib.pyplot as plt

averageTime = []
averageError = []
gridSize = [2, 4, 8, 16, 20, 25, 32]
```

```python
l1 = ['dynamicRow.txt', 'dynamicRowBatch.txt', 'dynamicRowChunk.txt', 'dynamicRowChunkBatch.txt']

for name in l1:
    with open(name, 'r') as f:
        data = f.read().split('\n')
    
    data = data[1:]

    avgTime = []
    avgError = []
    i = 0
    while (i<len(data)-1):
        i += 1
        avgTime.append(np.mean([float(data[j]) for j in range(i, i+7, 3)]))
        i += 1
        avgError.append(np.mean([float(data[j]) for j in range(i, i+7, 3)]))
        i += 7
    averageTime.append(avgTime)
    averageError.append(avgError)

plt.plot(gridSize, averageTime[0], 'o-', label='dynamicRow')
plt.plot(gridSize, averageTime[1], 'o-', label='dynamicRowBatch')
plt.plot(gridSize, averageTime[2], 'o-', label='dynamicRowChunk')
plt.plot(gridSize, averageTime[3], 'o-', label='dynamicRowChunkBatch')
plt.xlabel('Threads')
plt.ylabel('Average Time')
plt.title('Average Time vs Threads')
plt.legend()
plt.show()

plt.plot(gridSize, averageError[0], 'o-', label='dynamicRow')
plt.plot(gridSize, averageError[1], 'o-', label='dynamicRowBatch')
plt.plot(gridSize, averageError[2], 'o-', label='dynamicRowChunk')
plt.plot(gridSize, averageError[3], 'o-', label='dynamicRowChunkBatch')
plt.xlabel('Threads')
plt.ylabel('Average Error')
plt.title('Average Error vs Threads')
# increase y - limit to see the difference
plt.ylim(1.42e-6, 1.55e-6)
plt.legend()
```
